# About Simulations:

```python
from Simulation import Simulation
from backfire_update_funs import BF_Update_Functions, NewUpdate
from belief_states import build_belief, Belief
from influence_graphs import Influence,build_influence
import matplotlib.pyplot as plt
```

```python
#Creating function container:
fun_container=BF_Update_Functions()
```

```python
#Function to plot results
def plot_simulation(sim,title,max_time=100):
    plt.plot(sim[1][:max_time])
    plt.title(title)
    plt.legend(loc="upper right") 
    plt.show()
```

```python
#Line function with k=0.5, circular influence and triple initial belief, 50 agents
duration=2000
lineSimRaw=Simulation(
        build_belief(Belief.TRIPLE,50),
        build_influence(Influence.CIRCULAR,50),
        fun_container.get_function((NewUpdate.LINE,0.5))
    )
lineSim=lineSimRaw.run(duration,smart_stop=False)
plot_simulation(lineSim,"Line Function",duration)
```

```python
#Modulus function with k=-0.5, circular influence and extreme initial belief, 100 agents
duration=1000
lineSimRaw=Simulation(
        build_belief(Belief.EXTREME,50),
        build_influence(Influence.CIRCULAR,50),
        fun_container.get_function((NewUpdate.MODULUS,-0.5))
    )
lineSim=lineSimRaw.run(duration,smart_stop=True)
plot_simulation(lineSim,"Modulus Function",duration)
```
